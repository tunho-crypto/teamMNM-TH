```python
points = [
    [35.6895,139.6917],
    [34.0522,-118.2437],
    [51.5074,-0.1278],
    [48.8566,2.3522],
]

#Tính tâm
centroid_lat = sum([point[0] for point in points])/len(points)
centroid_lon = sum([point[1] for point in points])/len(points)
centroid = [centroid_lat,centroid_lon]
print("Tâm của các điểm là:",centroid)
```

```python
#list : là chuỗi có thứ tự có thẩy thay đổi , thêm sửa xóa , có thể trùng
path = [
    (35.6895,139.6917),
    (34.0522,-118.2437),
    (51.5074,-0.1278),
]

path.append((48.8566,2.3522))
print("Path da duoc cap nhat:",path)
sub_path = path[:2] #cắt 1 vị trí thứ 2 của biến path
print("Sub-path:",sub_path)
```

```python
#set là tập hợp các phần tử duy nhất ko theo thứ tự,ko lưu các giá trị trùng nhau
regions = ["North America","Europe","Asia"]
regions = set(regions)
#Thêm một khu vực mới
regions.add("Africa")
print("Khu vực được cập nhật",regions)

regions.add("Europe")
print("Ket qua sau khi them du lieu trung",regions)
```

```python
#dictionary : tập hợp các cặp khóa giá trị
city_attributes = {
    "name":"Tokyo",
    "population":"13929286",
    "coordinates":(35.6895,139.6917),
}

city_name = city_attributes["name"]
city_population = city_attributes["population"]
print(f"City:{city_name},Population:{city_population}")

city_attributes["area_km2"] = 2191 # thêm diện tích của thành phố km2
print("Thuộc tính của thành phố đã được cập nhật:",city_attributes)
```

```python
#Bài tập 1
cities_coords = [
    (21.0285, 105.8542),  # Hà Nội
    (16.0544, 108.2022),  # Đà Nẵng
    (10.8231, 106.6297)   # TP.HCM
]

centroid_lat = sum([city[0] for city in cities_coords])/len(cities_coords)
centroid_lon = sum([city[1] for city in cities_coords])/len(cities_coords)

centroid = {
    "Vĩ độ tâm": centroid_lat,
    "Kinh độ tâm": centroid_lon
}
print("Tâm của các thành phố là:", centroid)
```
