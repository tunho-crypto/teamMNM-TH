```python
import random
```

```python
#vòng lặp for cho phép lặp lại 1 chuỗi(list,tuple,string)
coordinates = [
    (35.6895, 139.6917),
    (34.0522, -118.2437),
    (51.5074, -0.1278),
]
for lat, lon in coordinates:
    print(f"Latitude:{lat},Longitude:{lon}")
```

```python
def calculate_distance(lat1, lon1, lat2, lon2):
    return ((lat2 - lat1) ** 2 + (lon2 - lon1) ** 2) ** 0.5
reference_point = (0, 0)
for lat, lon in coordinates:
    distance = calculate_distance(reference_point[0], reference_point[1], lat, lon)
    print(f"Distance from {reference_point} to ({lat},{lon}):{distance:.2f}")
```

```python
#vong lap while
counter = 0
while counter < len(coordinates):
    lat, lon = coordinates[counter]
    print(f"Processing coordinate:({lat}, {lon})")
    counter += 1
```

```python
for lat, lon in coordinates:
    if lat > 0 :
        print(f"{lat} is in the Northern Hemisphere")
    elif lat < 0 :
        print(f"{lat} is in the Southern Hemisphere")
    else :
        print(f"{lat} is near the equator")
```

```python
for lat, lon in coordinates:
    if lat > 0 :
        hemisphere = "Northern"
    else :
        hemisphere = "Southern"
    if lon > 0 :
        direction = "Eastern"
    else :
        direction = "Western"
    print(
        f"The coordinate ({lat}, {lon}) is in the {hemisphere} Hemisphere and { direction} Hemisphere."
    )
```

```python
filtered_coordinates = []
for lat, lon in coordinates:
    if lon > 0 :
        filtered_coordinates.append((lat, lon))
print(f"Filtered coodinates (only with positive longitude): {filtered_coordinates}")
```

```python
southern_count = 0
for lat, lon in coordinates :
    if lat < 0 :
        southern_count += 1
print(f"Number of coordinates inthe Southern Hemisphere: {southern_count}")
```

```python
cities = [
    {"name": "Hà Nội", "lat": 21.0285, "lon": 105.8542},
    {"name": "Hải Phòng", "lat": 20.8449, "lon": 106.6881},
    {"name": "Đà Nẵng", "lat": 16.0544, "lon": 108.2022},
    {"name": "TP. HCM", "lat": 10.8231, "lon": 106.6297},
    {"name": "Cần Thơ", "lat": 10.0452, "lon": 105.7469}
]
print("BÀI 1: CÁC THÀNH PHỐ MIỀN BẮC ")
for city in cities:
    # Giả sử vĩ độ > 18 là miền Bắc
    if city["lat"] > 18:
        print(f"{city['name']} nằm ở miền Bắc với tọa độ: ({city['lat']}, {city['lon']})")

print("\nBÀI 2: DÙNG WHILE (Dừng khi vĩ độ < 11) ")
counter = 0
while counter < len(cities):
    city = cities[counter]
    print(f"Đang xét: {city['name']} ({city['lat']}, {city['lon']})")
    if city["lat"] < 11:
        print("-> Đã tìm thấy tọa độ có vĩ độ nhỏ hơn 11. Dừng chương trình!")
        break    
    counter += 1

print("\nBÀI 3: ĐẾM VÀ IN CÁC TP MIỀN TRUNG & BẮC ")
count = 0 
for city in cities:
    # Nếu vĩ độ > 11 thì coi như là từ miền Trung trở ra Bắc
    if city["lat"] > 11:
        count += 1
        print(f"{city['name']}: ({city['lat']}, {city['lon']})")
print(f"Tổng cộng có {count} thành phố thuộc miền Trung và Bắc.")

print("\nBÀI 4: TẠO TỌA ĐỘ NGẪU NHIÊN ")
lat = random.uniform(0, 100)
lon = random.uniform(0, 100)  
print(f"Tọa độ sinh ra: ({lat:.2f}, {lon:.2f})")
while lat > 50 and lon > 50:
    lat = random.uniform(0, 100)
    lon = random.uniform(0, 100)    
    print(f"Tọa độ sinh ra: ({lat:.2f}, {lon:.2f})")
    if lat > 50 and lon > 50:
        print("-> Đã tìm thấy tọa độ thỏa mãn điều kiện. Dừng lại!")
        break 
```
